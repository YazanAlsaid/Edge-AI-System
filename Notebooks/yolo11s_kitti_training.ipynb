#4) Training the model

```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")
```

```python
yaml_path = "/content/drive/MyDrive/kitti_yolo/data.yaml"

results = model.train(
    data=yaml_path,
    epochs=150,
    imgsz=640,
    device=0,
    patience=15,
    name="yolo11s_kitti",
    project="/content/drive/MyDrive/yolo11s_kitti_runs"
)
```

```python
import random
import glob
import matplotlib.pyplot as plt
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/yolo11s_kitti_runs/yolo11s_kitti/weights/best.pt")

img_dir = "/content/drive/MyDrive/kitti_yolo/images/val"


all_images = glob.glob(img_dir + "/*.png") + glob.glob(img_dir + "/*.jpg")
sample_images = random.sample(all_images, 10)

# 4. 10 Bilder durch das Modell schicken
for idx, img_path in enumerate(sample_images):
    results = model.predict(img_path, imgsz=640, conf=0.25)

    print(f"\nğŸ”¹ Bild {idx+1}: {img_path}")
    plt.figure(figsize=(10, 6))
    plt.imshow(results[0].plot())
    plt.axis("off")
    plt.show()
```
